# Gas Report Analysis (Grouped by Description)

```python
import pandas as pd

# Load your gas report file (update path if needed)
file_path = "gas-report.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Group by 'Description' and compute average gas used
avg_gas_used = df.groupby("Description")["Gas Used"].mean().reset_index()
avg_gas_used.rename(columns={"Gas Used": "Average Gas Used"}, inplace=True)
avg_gas_used
```

```python
avg_gas_used.to_excel("average-gas-used-by-description.xlsx", index=False)
```
